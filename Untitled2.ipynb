```python
import pandas as pd
```

Label encoding : Label Encoding is a preprocessing technique used in machine learning to convert categorical (text) data into numerical form so that algorithms can understand it.

```python
df = pd.DataFrame({"name" : ["vicky" , "cow" , "taksh" , "venu" , "sunil"]})
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df["en_name"] = le.fit_transform(df["name"])
```

```python
dataset = pd.read_csv("loan.csv")
```

```python
la = LabelEncoder()
la.fit(dataset["Property_Area"])
```

```python
dataset["Property_Area"] = la.transform(dataset["Property_Area"])
```
