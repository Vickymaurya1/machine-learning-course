```python
import pandas as pd
```

One Hot encoding : One-hot encoding is a way to convert categorical (text) data into numerical form so that machines can understand it.

```python
dataset = pd.read_csv("loan.csv")
```

```python
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
#dataset["Gender"].fillna(dataset["Gender"].mode()[0], inplace = True)
dataset.fillna({"Married": dataset["Married"].mode()[0]}, inplace=True)
```

```python
en_data = dataset[["Gender" , "Married"]]
```

```python
pd.get_dummies(en_data)
```

```python
pd.get_dummies(en_data).info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop = "first")
ar = ohe.fit_transform(en_data).toarray()
ar
```

```python
pd.DataFrame(ar, columns = ["Gender_Male"  , "Married_Yes" ])
```
