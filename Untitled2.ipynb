```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("loan.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.fillna({"ApplicantIncome": dataset["ApplicantIncome"].mode()[0]}, inplace=True)
```

```python
sns.distplot(dataset["ApplicantIncome"])
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(dataset[["ApplicantIncome"]])
```

```python
dataset["ApplicantIncome_ss"] = pd.DataFrame(ss.transform(dataset[["ApplicantIncome"]]) , columns = ["x"])
```

```python
dataset.head(3)
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(10,7))
plt.subplot(1,2,1)
plt.title("Before")
sns.displot(dataset["ApplicantIncome_ss"])

plt.subplot(1 , 2 , 2)
plt.title("After")
sns.distplot(dataset["ApplicantIncome"])

plt.show()
```
