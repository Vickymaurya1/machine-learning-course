Handling missing values using scikit learn

```python
dataset = pd.read_csv(r"/content/loan.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
dataset.select_dtypes(include = "float64").columns
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer(strategy = "mean")
ar = si.fit_transform(dataset[['CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term',
       'Credit_History']])
```

```python
new_dataset = pd.DataFrame(ar, columns = dataset.select_dtypes(include = "float64").columns)
```

```python
new_dataset.isnull().sum()
```
